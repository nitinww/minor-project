```python
import pandas as pd
```

```python
companies = pd.read_csv("companies.csv")
```

```python
tickers = companies.iloc[:, 0].tolist()
urls = companies.iloc[:, 1].tolist()
```

```python
for i in range(len(tickers)):
    data = pd.DataFrame()
    for j in range(17, 0, -1):
        url = "https://www.moneycontrol.com/financials//results/quarterly-results/" + urls[i] + "/" + str(j)
        # print(url)
        try:
            df = pd.read_html(url)[0]
            df = df.iloc[:, :-1]
            df.columns = df.iloc[0]
            df = df.drop(0)
            df = df.reset_index(drop=True)
            df = df.drop(0)
            df.columns.values[0] = None
            df.set_index(df.columns[0], inplace=True)
            df = df.iloc[:, ::-1]
            data = pd.concat([data, df], axis=1)
        except Exception as e:
            print(f"Error processing data for {tickers[i]} at URL {url}: {e}")
    
    if not data.empty:
        data = data.T
        path = "../datasets/quarterly-data/" + tickers[i] + "-quarterly-data.csv"
        data.to_csv(path)
        print(f"CSV file created for {tickers[i]}")
    else:
        print(f"No data available for {tickers[i]}")
```
