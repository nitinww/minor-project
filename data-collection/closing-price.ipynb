```python
import pandas as pd
import yfinance as yf
```

```python
ticker = "WIPRO"
```

```python
df = pd.read_csv(f'../datasets/{ticker}-without-closing-price.csv', index_col=False)
df.set_index(df.columns[0], inplace=True)
df
```

```python
ticker = yf.download(f'{ticker}.NS', start="2004-03-01", end="2024-10-01", interval="1mo")

ticker.index = pd.to_datetime(ticker.index)

ticker = ticker[ticker.index.month.isin([3, 6, 9, 12])]
```

```python
ticker.shape
```

```python
ticker = ticker.T
```

```python
ticker.drop(['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
```

```python
ticker.columns = df.columns
```

```python
ticker
```

```python
df = pd.concat([ticker, df])
```

```python
df
```

```python
df.to_csv(f"../datasets/not-final/WIPRO.csv")
```
